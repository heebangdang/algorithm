```python
import pandas as pd
data = pd.read_csv("nonstoprocketdata.csv")
data
```

```python
a = 0.25*data["Fatality"] + 0.55*data["patient"]+ 0.2*data["medical level"]
data["value"] = a
df1 = data[["Continent","Country","value","Priority"]]
df1
```

```python
countrynumber = int(input("How many country do you want?"))
countrylist = []
for i in range(countrynumber) :
    country = input("input country where you want :")
    countrylist.append(country)
df2 = df1[df1["Country"].isin(countrylist)]
df2
```

```python
import numpy
d = {}
for i in df2['Country'].unique():
    d[i] = [df2['Continent'][j] for j in df2[df2['Country']==i].index]
d
dlist = []
for key in d.keys():
    dlist.append(d[key])
dlist = numpy.unique(dlist).tolist()
dlist
```

```python
data = [[0.5,-1]]
for i in range(len(dlist)):
    if dlist[i] == 'North America':
        data.append([1.5,-0.4])
    elif dlist[i] == 'South America':
        data.append([1.8,-3])
    elif dlist[i] == 'Oceania':
        data.append([0.6,-3.6])
    elif dlist[i] == 'Africa':
        data.append([-1,-2.3])
    elif dlist[i] == 'Europe':
        data.append([-0.8,0.1])
    elif dlist[i] == 'Middle East':
        data.append([-0.3,-1])
    elif dlist[i] == 'Asia':
        data.append([0,0])
    i=i+1
print(data)

s=data[0]
sequence=['Korea']
for r in range(len(data)):
    l=[]
    for p in range(len(data)-1):
        x=s
        y=data[p+1]
        z=[i-j for i,j in zip(x,y)]
        c=0
        for k in range(len(z)):
            c+=(z[k])**2
            k+=1
        l.append(c**(1/2))
    print(l)
    
    while len(l)>0:
        smallest=l[0]
        for q in l:
            if q < smallest:
                smallest = q
                q+=1
        print(smallest)
        
        for n in range(len(l)):
            if smallest == l[n]:
                s=data[n]
                del(data[n+1])
                sequence.append(dlist[n])
                del(dlist[n])
        break
    print(sequence)
    print(data)
    
if len(l)==0:
    sequence.append('Korea')
print("The sequence of continent is",sequence)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np 

posn=[]
for i in range(len(sequence)) :
    posn.append(sequence[i])

Con=nx.DiGraph()

Con.add_nodes_from(["{}".format('Korea')])
Con.add_nodes_from(["{}".format(sequence[i]) for i in range(len(sequence))])
Con.add_edges_from([("{}".format(posn[i]), "{}".format(posn[i+1])) for i in range(len(posn)-1)])

pos = {'Korea':[0.5, -1], 'North America':[1.5, -0.4], 'South America':[1.8, -3], 'Oceania':[0.6, -3.6], 
       'Africa':[-1, -2.3], 'Europe':[-0.8, 0.1], 'Middle East':[-0.3, -1], 'Asia':[0, 0]}
      
plt.figure(figsize=(20, 12))
plt.ylim(-4.0, 1.0)

nx.draw_networkx_nodes(Con, pos, node_size=6000, alpha=1.0, node_shape='s', node_color='pink')
nx.draw_networkx_edges(Con, pos, width=2, alpha=0.8, edge_color='g')
#arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")

labels = {}
for idx, node in enumerate(Con.nodes()):
    labels[node] = '{}'.format(posn[idx])
nx.draw_networkx_labels(Con, pos, labels, font_size=15)
 
plt.axis('off')
plt.tight_layout()
plt.show()

print("The sequence is",sequence)
```

# Time Comparison of Sorting Method

```python
from time import time
import numpy as np
import pandas as pd

x = [10000,20000,30000,40000]

testTimeOne  =[]
a=0
suma=0
for i in x:
    for j in range(1000) :
        np.random.seed(10)
        list_colum=['Fatality','patient','medical level']
        ran_df_1 = pd.DataFrame(np.random.randint(1,4,size=(i, 3)), columns=list_colum)
        ran_df_1["Country"] = range(1, i+1)
        start_time = time()
        std_1 = ran_df_1.sort_values(by=["Fatality"], kind='heapsort')
        end_time = time()
        t = end_time - start_time
        a+=t
    suma=a/1000
    testTimeOne.append(suma)
    
    
testTimeTwo  =[]
b=0
sumb=0
for i in x:
    for j in range(1000) :
        np.random.seed(10)
        list_colum=['Fatality','patient','medical level']
        ran_df_2 = pd.DataFrame(np.random.randint(1,4,size=(i, 3)), columns=list_colum)
        ran_df_2["Country"] = range(1, i+1)
        start_time = time()
        std_2 = ran_df_2.sort_values(by=["Fatality","patient"], kind='heapsort')
        end_time = time()
        t = end_time - start_time
        b+=t
    sumb=b/1000
    testTimeTwo.append(sumb)                                                                                                                                                                                                                                                                                     
    
testTimeThree  =[]
c=0
sumc=0
for i in x:
    for j in range(1000) :
        np.random.seed(10)
        list_colum=['Fatality','patient','medical level']
        ran_df_3 = pd.DataFrame(np.random.randint(1,4,size=(i, 3)), columns=list_colum)
        ran_df_3["Country"] = range(1, i+1)
        start_time = time()
        std_3 = ran_df_3.sort_values(by=["Fatality","patient","medical level"], kind='heapsort')
        end_time = time()
        t = end_time - start_time
        c+=t
    sumc=c/1000
    testTimeThree.append(sumc)
    
testTimePriority  =[]
d=0
sumd=0
for i in x:
    for j in range(1000) :
        np.random.seed(10)
        list_colum=['Fatality','patient','medical level']
        ran_df_4 = pd.DataFrame(np.random.randint(1,4,size=(i, 3)), columns=list_colum)
        ran_df_4["Country"] = range(1, i+1)
        start_time = time()
        ran_df_4["priority"] = 0.25*ran_df_4["Fatality"] + 0.55*ran_df_4["patient"]+ 0.2*ran_df_4["medical level"]
        std_4 = ran_df_4.sort_values(["priority"], kind='heapsort')
        end_time = time()
        t = end_time - start_time
        d+=t
    sumd=d/1000
    testTimePriority.append(sumd)

print(testTimeOne)
print(testTimeTwo)
print(testTimeThree) 
print(testTimePriority)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab

pylab.xlabel('Data')
pylab.ylabel('Time (sec)')

plt.title("One attribute Sort vs Two attribute sort vs Three attribute sort vs Priority Sort")
plt.plot(x, testTimeOne, "ro-", c="red")
plt.plot(x, testTimeTwo, "ro-", c="blue")
plt.plot(x, testTimeThree, "ro-", c="orange")
plt.plot(x, testTimePriority, "ro-", c="green")
plt.legend(["One attribute Sort", "Two attribute Sort","Three attribute Sort","Priority Sort"], loc='upper left')
plt.show()
```
